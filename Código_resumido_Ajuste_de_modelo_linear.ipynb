# **Ajuste de uma equação da reta a partir de pontos experimentais** 

```
#Entrada de dados
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2.2, 4.1, 6.7, 8.0, 9.8, 13]
#Estimando o modelo
import numpy as np
mod_linear = np.polyfit(x, y, 1)
novo_y = np.polyval(mod_linear, x)
#Construindo o gráfico
import matplotlib.pyplot as plt
plt.plot(x,y, "o", label = "pontos experimentais")
plt.plot(x,novo_y, "-r", label = "modelo")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajuste de uma reta")
plt.legend()
plt.show()
#Estimando o coeficiente de determinação
yresid = y - novo_y
SQresid = sum(yresid**2)
y_tot = y - np.mean(y)
SQtotal = sum(y_tot**2)
R2 = 1 - SQresid/SQtotal
print('O coeficiente de determinação do ajuste é {:.4f}'.format(R2))
```
